## Standard Scaler

The Standard Scaler is a preprocessing technique in machine learning used to normalize or standardize the features of a dataset. It is part of the sklearn.preprocessing module in the scikit-learn library.

Purpose of Standard Scaler:
The main goal of the Standard Scaler is to transform features so they have:

Mean = 0 (Centering the data)
Standard Deviation = 1 (Scaling the data)

This ensures that all features contribute equally to the model's learning process and are not influenced by their scale or magnitude.

Why Use Standard Scaler in Machine Learning?
Improves Model Performance: Algorithms like Support Vector Machines (SVM), k-Nearest Neighbors (k-NN), Principal Component Analysis (PCA), and Gradient Descent-based models (e.g., Logistic Regression) are sensitive to the scale of data. Standardizing features can lead to faster convergence and better accuracy.

Avoids Dominance of Larger-Scale Features: If features have different scales (e.g., one in the range 0-1, another in the range 0-1000), models might give more weight to the larger-scale feature, leading to biased predictions.

Ensures Optimal Regularization: For models like Ridge and Lasso regression, which penalize large coefficients, standardization ensures the penalty terms are applied consistently across all features.

Implementation in Python:
from sklearn.preprocessing import StandardScaler
import numpy as np

# Sample data
data = np.array([[1.0, 2.0], [2.0, 4.0], [3.0, 6.0]])

# Initialize scaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(data)

print("Original Data:\n", data)
print("Scaled Data:\n", scaled_data)

Output:
Original Data:

[[1. 2.]
 [2. 4.]
 [3. 6.]]

Scaled Data:

[[-1.22474487 -1.22474487]
 [ 0.          0.        ]
 [ 1.22474487  1.22474487]]

Key Notes:
Use the Standard Scaler on training data and apply the same transformation to the test data using the fitted scaler (fit_transform on training data, transform on test data).
It's most effective when features are approximately normally distributed.


```python
from sklearn.preprocessing import StandardScaler
```

## train_test_split

```python
from sklearn.model_selection import train_test_split
input_data_train, input_data_test, output_data_train, output_data_test = train_test_split( input_data, output_data, test_size = 0.3)
```

```python
print(input_data_train.shape)
print(output_data_train.shape)
print(input_data_test.shape)
print(output_data_test.shape)
```

## Buliding Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
# Initialize the  model
model = LinearRegression()
```

```python
# Train the model
model.fit(input_data_train, output_data_train)
```

```python
#Prediction
predicted_sellingprice = model.predict(input_data_test)
```

## Evalution Matrix of Linear Regression

```python
from sklearn import metrics
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('Mean Absolute Error:', (metrics.mean_absolute_error(predicted_sellingprice, output_data_test)))
print('Mean Squared Error:', (metrics.mean_squared_error(predicted_sellingprice, output_data_test)))
print('R2 Score:', (metrics.r2_score(predicted_sellingprice, output_data_test)))
```
